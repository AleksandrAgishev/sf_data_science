```python
age = 2018 - bike['birth year']
display(age)
```

```python
bike['age'] = age
bike.drop(['birth year'],axis=1, inplace=True)
display(bike)
```

```python
display(bike[bike['age']>60]['age'].count())
```

```python
trip_duration = pd.to_datetime(bike['stoptime'])-pd.to_datetime(bike['starttime'])
display(trip_duration.dt.seconds/60)
```

```python
weekend=pd.to_datetime(bike['starttime']).dt.dayofweek
bike['weekend'] = weekend.apply(lambda x: 1 if (x==5 or x==6) else 0)
display(bike['weekend'].value_counts())

```

```python
time_of_day = pd.to_datetime(bike['starttime']).dt.hour
def tod (hours):
    if hours >=0 and hours <=6:
        return 'night'
    elif hours <=12:
        return 'morning'
    elif hours <=18:
        return'day'
    else:
        return 'evening'

bike['time_of_day'] = time_of_day.apply(tod)
display(bike['time_of_day'].value_counts())
```
